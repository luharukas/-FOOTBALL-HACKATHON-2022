```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data=pd.read_csv('train_df.csv')
test_data=pd.read_csv('test_df.csv')
```

```python
def scaling(x):
    if x==0:
        return 0.5
    elif x==1:
        return 0
    elif x==2:
        return 1

train_data['winner']=train_data['winner'].apply(scaling)
test_data['winner']=test_data['winner'].apply(scaling)
```

```python
train_data['player_position_1']=train_data['player_position_1'].apply(np.round).astype('int')
train_data['player_position_2']=train_data['player_position_2'].apply(np.round).astype('int')
train_data['team1_system_id']=train_data['team1_system_id'].apply(np.round).astype('int')
train_data['team2_system_id']=train_data['team2_system_id'].apply(np.round).astype('int')

test_data['player_position_1']=test_data['player_position_1'].apply(np.round).astype('int')
test_data['player_position_2']=test_data['player_position_2'].apply(np.round).astype('int')
test_data['team1_system_id']=test_data['team1_system_id'].apply(np.round).astype('int')
test_data['team2_system_id']=test_data['team2_system_id'].apply(np.round).astype('int')
```

```python
train_data.drop(columns=['player_var','team1_var','team2_var'],axis=1,inplace=True)
test_data.drop(columns=['player_var','team1_var','team2_var'],axis=1,inplace=True)
train_data
```

```python
raw_data=pd.read_csv(r'Data\train.csv')
```

```python
raw_test_data=pd.read_csv(r'Data\test.csv')
```

```python
train_player_encoded_data=raw_data.filter(regex='player_[a-z_]+var_[0-9]+')
train_team1_encoded_data=raw_data.filter(regex='team1_[a-z_]+var_[0-9]+')
train_team2_encoded_data=raw_data.filter(regex='team2_[a-z_]+var_[0-9]+')

test_player_encoded_data=raw_test_data.filter(regex='player_[a-z_]+var_[0-9]+')
test_team1_encoded_data=raw_test_data.filter(regex='team1_[a-z_]+var_[0-9]+')
test_team2_encoded_data=raw_test_data.filter(regex='team2_[a-z_]+var_[0-9]+')
```

```python
for i in train_player_encoded_data.columns:
    print(i,train_player_encoded_data[i].isnull().sum())
```

```python
import matplotlib.pyplot as plt
array={'columns':[],'missing_value':[]}
for i in train_player_encoded_data.columns:
    array['columns'].append(i)
    array['missing_value'].append(train_player_encoded_data[i].isnull().sum())
    
array=pd.DataFrame(array)
plt.scatter(array['columns'],array['missing_value'])
plt.show()
```

```python
null_rows=train_player_encoded_data[train_player_encoded_data['player_offensive_raw_var_31'].isnull()].index.tolist()
train_player_encoded_data.drop(index=null_rows,axis=0,inplace=True)
len(train_player_encoded_data)
```

```python
for i in train_player_encoded_data.columns:
    if train_player_encoded_data[i].isnull().sum() >=15000:
        train_player_encoded_data.drop(columns=i,axis=1,inplace=True)
        test_player_encoded_data.drop(columns=i,axis=1,inplace=True)
    else:
        mean_value=train_player_encoded_data[i].mean()
        train_player_encoded_data[i].fillna(value=mean_value, inplace=True)
        test_player_encoded_data[i].fillna(value=mean_value, inplace=True)
```

```python
player_offensive=train_player_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
player_defensive=train_player_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
player_positional=train_player_encoded_data.filter(regex='[a-z_]+positional[a-z0-9_]+')
player_physical=train_player_encoded_data.filter(regex='[a-z_]+physical[a-z0-9_]+')
player_general=train_player_encoded_data.filter(regex='[a-z_]+general[a-z0-9_]+')
player_other=train_player_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')


test_player_offensive=test_player_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
test_player_defensive=test_player_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
test_player_positional=test_player_encoded_data.filter(regex='[a-z_]+positional[a-z0-9_]+')
test_player_physical=test_player_encoded_data.filter(regex='[a-z_]+physical[a-z0-9_]+')
test_player_general=test_player_encoded_data.filter(regex='[a-z_]+general[a-z0-9_]+')
test_player_other=test_player_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')
```

```python
train_player_encoded_data['player_offensive']=np.average(player_offensive,axis=1)
train_player_encoded_data['player_defensive']=np.average(player_defensive,axis=1)
train_player_encoded_data['player_positional']=np.average(player_positional,axis=1)
train_player_encoded_data['player_physical']=np.average(player_physical,axis=1)
train_player_encoded_data['player_general']=np.average(player_general,axis=1)
train_player_encoded_data['player_other']=np.average(player_other,axis=1)


test_player_encoded_data['player_offensive']=np.average(test_player_offensive,axis=1)
test_player_encoded_data['player_defensive']=np.average(test_player_defensive,axis=1)
test_player_encoded_data['player_positional']=np.average(test_player_positional,axis=1)
test_player_encoded_data['player_physical']=np.average(test_player_physical,axis=1)
test_player_encoded_data['player_general']=np.average(test_player_general,axis=1)
test_player_encoded_data['player_other']=np.average(test_player_other,axis=1)
```

```python
for i in train_team1_encoded_data.columns:
    if train_team1_encoded_data[i].isnull().sum() >=15000:
        train_team1_encoded_data.drop(columns=i,axis=1,inplace=True)
        test_team1_encoded_data.drop(columns=i,axis=1,inplace=True)
    else:
        mean_value=train_team1_encoded_data[i].mean()
        train_team1_encoded_data[i].fillna(value=mean_value, inplace=True)
        test_team1_encoded_data[i].fillna(value=mean_value, inplace=True)
```

```python
for i in train_team2_encoded_data.columns:
    if train_team2_encoded_data[i].isnull().sum() >=15000:
        train_team2_encoded_data.drop(columns=i,axis=1,inplace=True)
        test_team2_encoded_data.drop(columns=i,axis=1,inplace=True)
    else:
        mean_value=train_team2_encoded_data[i].mean()
        train_team2_encoded_data[i].fillna(value=mean_value, inplace=True)
        test_team2_encoded_data[i].fillna(value=mean_value, inplace=True)
```

```python
team1_offensive=train_team1_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
team1_defensive=train_team1_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
team1_other=train_team1_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')

team2_offensive=train_team2_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
team2_defensive=train_team2_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
team2_other=train_team2_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')

test_team1_offensive=test_team1_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
test_team1_defensive=test_team1_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
test_team1_other=test_team1_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')

test_team2_offensive=test_team2_encoded_data.filter(regex='[a-z_]+offensive[a-z0-9_]+')
test_team2_defensive=test_team2_encoded_data.filter(regex='[a-z_]+defensive[a-z0-9_]+')
test_team2_other=test_team2_encoded_data.filter(regex='[a-z_]+other[a-z0-9_]+')
```

```python
train_team1_encoded_data['team1_offensive']=np.average(team1_offensive,axis=1)
train_team1_encoded_data['team1_defensive']=np.average(team1_defensive,axis=1)
train_team1_encoded_data['team1_other']=np.average(team1_other,axis=1)
train_team2_encoded_data['team2_offensive']=np.average(team2_offensive,axis=1)
train_team2_encoded_data['team2_defensive']=np.average(team2_defensive,axis=1)
train_team2_encoded_data['team2_other']=np.average(team2_other,axis=1)


test_team1_encoded_data['team1_offensive']=np.average(test_team1_offensive,axis=1)
test_team1_encoded_data['team1_defensive']=np.average(test_team1_defensive,axis=1)
test_team1_encoded_data['team1_other']=np.average(test_team1_other,axis=1)
test_team2_encoded_data['team2_offensive']=np.average(test_team2_offensive,axis=1)
test_team2_encoded_data['team2_defensive']=np.average(test_team2_defensive,axis=1)
test_team2_encoded_data['team2_other']=np.average(test_team2_other,axis=1)
```

```python
from copy import deepcopy
from math import log,pow,sqrt,exp
def sqr(x):
    return x**2
def sqrt(x):
    return pow(x,1/2)
def inv(x):
    return 1/x
def ex(x):
    return exp(x)
def lo(x):
    return log(x)
def simple(x):
    return x
temp_final_data=deepcopy(final_data)
temp_final_data['player_defensive']=temp_final_data['player_defensive'].apply(log)
temp_final_data['player_positional']=temp_final_data['player_positional'].apply(log)
temp_final_data['player_physical']=temp_final_data['player_physical'].apply(sqr)
temp_final_data['player_general']=temp_final_data['player_general'].apply(log)

temp_final_data['team1_offensive']=temp_final_data['team1_offensive'].apply(sqrt)
temp_final_data['team1_defensive']=temp_final_data['team1_defensive'].apply(log)
temp_final_data['team2_offensive']=temp_final_data['team2_offensive'].apply(sqrt)
temp_final_data['team2_defensive']=temp_final_data['team2_defensive'].apply(log)
```

```python
import scipy.stats as stats
import pylab
def normality(data,feature):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()

for i in ['team2_defensive']:
    normality(temp_final_data,i)
```

# Scaling of the parameter

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler

scaler1 = MinMaxScaler()
scaler2 = StandardScaler()
final_data.iloc[:,:-1]=scaler1.fit_transform(final_data.iloc[:,:-1])
final_test_data.iloc[:,:]=scaler1.fit_transform(final_test_data.iloc[:,:])
```

```python
from sklearn.model_selection import train_test_split
xtrain=final_data.iloc[:,:-1]
ytrain=final_data.iloc[:,-1:]
xtest=final_test_data.iloc[:,:]
```

```python
xtest.shape
```

# lbgm

```python
import lightgbm
from sklearn.model_selection import GridSearchCV
lbgm_boost={'max_depth':[10],'n_estimators':[1000],'boosting_type':['gbdt','dart'],'learning_rate':[0.01,0.05,0.1]}
base_estimator=lightgbm.LGBMRegressor(random_state=42)
model=GridSearchCV(estimator=base_estimator,param_grid=lbgm_boost,scoring='r2',cv=5,)
model.fit(xtrain,ytrain,)
```

```python
print('Best score for training data:', model.best_score_,"\n") 


print('Best Booster:',model.best_estimator_.boosting_type,"\n") 
print('Best Learning rate:',model.best_estimator_.learning_rate,"\n")
print('Best Max_depth:',model.best_estimator_.max_depth,"\n") 
print('Best n_estimators:',model.best_estimator_.n_estimators,"\n")
final_model = model.best_estimator_
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
params={'max_iter':[10000],'warm_start':[True,False],'alpha':[0.01,0.001,0.05]}

estimator=Lasso(random_state=42)
model=GridSearchCV(estimator=estimator,param_grid=params,scoring='r2',cv=5)
model.fit(xtrain,ytrain)

final_model=model.best_estimator_
print(model.best_estimator_.alpha)
print(model.best_estimator_.max_iter)
print(model.best_estimator_.warm_start)
print(model.best_score_)
```

# Random Regressor

```python
from sklearn.ensemble import RandomForestRegressor
grid = { 
    'n_estimators': [300],
    'max_depth' : [10],
}
## Grid Search function
model = GridSearchCV(estimator=RandomForestRegressor(random_state=42,), param_grid=grid, cv=5,scoring='r2')
model.fit(xtrain,ytrain)
```

```python
print('Best score for training data:', model.best_score_,"\n") 
print('Best estimator:',model.best_estimator_.n_estimators,"\n") 
print('Best max_depth:',model.best_estimator_.max_depth,"\n")
final_model = model.best_estimator_
```

# XGBooster

```python
from xgboost import XGBRegressor
grid_xgboost={'n_estimators':[500],'max_depth':[10],'booster':['gbtree', 'dart'],'learning_rate':[0.1,0.05]}
base_estimator=XGBRegressor(random_state=42)
model=GridSearchCV(estimator=base_estimator,param_grid=grid_xgboost, cv=5,scoring='r2')
model.fit(xtrain,ytrain) 
```

```python
print('Best score for training data:', model.best_score_,"\n") 
print('Best Booster:',model.best_estimator_.booster,"\n") 
print('Best Learning rate:',model.best_estimator_.learning_rate,"\n")
print('Best Max_depth:',model.best_estimator_.max_depth,"\n") 
print('Best n_estimators:',model.best_estimator_.n_estimators,"\n")
final_model = model.best_estimator_
```
