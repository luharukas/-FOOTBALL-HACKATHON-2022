# Load Important libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
sns.set(rc = {'figure.figsize':(15,8)})
import warnings
warnings.filterwarnings('ignore')
```

# Load the data

```python
train_data_path='Data/train.csv'
test_data_path='Data/test.csv'
```

```python
def load_data(filepath):
    return pd.read_csv(filepath)

train_data=load_data(train_data_path)
test_data_path=load_data(test_data_path)
train_data.head(20)
```

```python
# relocation of "rating_num" column in the last
cols = train_data.columns.tolist()
cols=cols[:2]+cols[3:]+cols[2:3]
train_data=train_data[cols]
```

```python
# relation between scout_id and average of rating_num
from numpy import median,mean
ax = plt.subplots()

 
ax=sns.barplot(train_data['scout_id'],train_data['rating_num'],estimator=mean, color='blue')
ax=sns.barplot(train_data['scout_id'],train_data['rating_num'],estimator=median, color='green')

ax.set(xlabel="x-axis", ylabel="y-axis")
plt.show()

print("Correlation between scout_id and rating_num : ",train_data['scout_id'].corr(train_data['rating_num']))
```

<p> This is a symmmentrical data. Scout_id is not very much dependent on the rating_num. Its correlation value is very much low. </p>

```python
#one hot encoding of the winner column and decompose into winner, loser and draw columns
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore',dtype='int')
encoder_df = pd.DataFrame(encoder.fit_transform(train_data[['winner']],).toarray())
train_data = train_data.join(encoder_df)

train_data.drop('winner', axis=1, inplace=True)
```

```python
# relocation of one hot encoded columns column in the last
cols = train_data.columns.tolist()
cols=cols[:2]+cols[-3:]+cols[2:-3]
train_data=train_data[cols]
train_data.rename(columns = {0:'draw', 1:'loser',2:'winner'}, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_data['team']=le.fit_transform(train_data['team'])
```

```python
train_player_encoded_data=train_data.filter(regex='player_[a-z_]+var_[0-9]+')
train_team1_encoded_data=train_data.filter(regex='team1_[a-z_]+var_[0-9]+')
train_team2_encoded_data=train_data.filter(regex='team2_[a-z_]+var_[0-9]+')
```

```python
train_player_encoded_data.fillna(train_player_encoded_data.mean(),inplace=True)

from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=42,max_depth=10)
model.fit(train_player_encoded_data,train_data['rating_num'])
print("Dimension Reduction done")

```

```python
features=train_player_encoded_data.columns
importances=model.feature_importances_
indices = np.argsort(importances)[-50:]  # top 80 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
selected_encoded_player_features=train_player_encoded_data[features[indices]]
```

```python
train_team1_encoded_data.fillna(train_team1_encoded_data.mean(),inplace=True)
```

```python
for i in train_team1_encoded_data.columns:
    if train_team1_encoded_data[i].isna().sum()!=0:
        train_team1_encoded_data.drop(labels=[i],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=42,max_depth=10)
model.fit(train_team1_encoded_data,train_data['rating_num'])
print("Dimension Reduction done")
```

```python
features=train_team1_encoded_data.columns
importances=model.feature_importances_
indices = np.argsort(importances)[-100:]  # top 80 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
selected_encoded_team1_features=train_team1_encoded_data[features[indices]]
```

```python
train_team2_encoded_data.fillna(train_team2_encoded_data.mean(),inplace=True)
```

```python
for i in train_team2_encoded_data.columns:
    if train_team2_encoded_data[i].isna().sum()!=0:
        train_team2_encoded_data.drop(labels=[i],axis=1,inplace=True)
```

```python
train_team2_encoded_data.fillna(train_team2_encoded_data.mean(),inplace=True)

from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=42,max_depth=15)
model.fit(train_team2_encoded_data,train_data['rating_num'])
print("Dimension Reduction done")
```

```python
features=train_team2_encoded_data.columns
importances=model.feature_importances_
indices = np.argsort(importances)[-100:]  # top 80 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
